# Project Description

In this project some aspects of the income of a person in relation to his college major are analyzed. Hereto, several aspects are visualized for investigation.

The dataset used is the [college major dataset](https://raw.githubusercontent.com/fivethirtyeight/data/master/college-majors/recent-grads.csv) provided by FiveThirtyEight.

# Import Libaries

```python
import io
import os
import urllib.request

from IPython.display import display

import pandas as pd

import pandas.tools.plotting as pd_plot
import matplotlib.pyplot as plt
%matplotlib inline
```

# Set Global Varibles

```python
URL = 'https://raw.githubusercontent.com/fivethirtyeight/data/master/college-majors/recent-grads.csv'
PATH_DATASET = 'data/' + URL.split('/')[-1]
```

# Project Preparation

## Download the data

```python
def download_github_csv_data(url):
    """Download a csv file and stores it in the data folder of the project repository.

    Args:
        URL of the csv file

    Returns:
        None
    """
    ### Create data dir if not exts
    if not os.path.exists('data/'):
        os.makedirs('data/')
        
    file = urllib.request.urlopen(url)

    df = pd.read_csv(io.TextIOWrapper(file))

    filename = url.split('/')[-1]
    path = 'data/' + filename
    df.to_csv(path, header=True, index=False, sep=',')

"""
Downloads the data to the data folder of a local repository after you run it once you can uncomment this lines.
To prevent the code from downloading the data every time you run the code.
"""    
download_github_csv_data(URL)
```

## Load the data

```python
### Load the data as df in the var data
data = pd.read_csv(PATH_DATASET)
```
